# Life expectancy and GDP per country over time from WHO

### Import packages and data

```python
import pandas as pd
import missingno as msno
from IPython.display import display
import seaborn as sns
from matplotlib import pyplot as plt

df = pd.read_csv("all_data.csv")
```

### Inspect head, tail, shape, data formats

```python
display(df.head())
display(df.tail())
print(df.shape)
print(df.info())
```

### Check if there are any missing values

```python
msno.matrix(df)
```

### Get Descriptive Statistics

```python
display(df.describe())
display(df.Country.value_counts())
```

### Heatmap of correlation coefficients between all quantitative features (Pairwise Correlation Coefficient Matrix)

```python
display(df.corr()) 
sns.heatmap(df.corr(), annot=True)
plt.show()
plt.clf()
```

__There are only three pairings to consider__:<br><br>
__Life Expectancy vs Year__: It is known from domain knowledge that life expectancy has indeed gone up over time and between generations, the low r score here is probably due to the low amount of data points not being enough to show the association.<br>
__GDP vs Year__: The fact it has a higher r score than Life expectancy here seems to indicate that its increase over time is more observable. The effect can be observed in only a small slice of time.<br>
__GDP vs Life Expectancy__: Shows the most linear association. Money can buy better healthcare adding more years to life can be a hypothesis.

### Pairplot matrix with and without colorization by country

```python
sns.pairplot(df, hue = "Country")
plt.show()
plt.clf()
sns.pairplot(df)
plt.show()
plt.clf()
```

__There are only three pairings to consider__:<br><br>

__Life Expectancy vs Year__: Zimbabwe is the odd one out. The entire group should be analyzed separately from Zimbabwe for this pairing.<br>
__GDP vs Year__: US and China pop out from the group. Suggests analysis of US and China and the rest of the group as separate categories.<br>
__GDP vs Life Expectancy__: Once again Zimbabwe is the odd one out from the group. The entire group should be analyzed separately from Zimbabwe.<br>

## Further Stratification after visually identifying data clusters

```python
df_no_zimbabwe = df[df.Country != 'Zimbabwe']
print(df_no_zimbabwe.Country.unique())
df_no_China_US = df[(df.Country != 'China') & (df.Country != 'United States of America')]
print(df_no_China_US.Country.unique())
```

```python
fig = plt.figure(figsize = (12, 8))
sns.lineplot(
    data=df_no_zimbabwe,
    x="Year", y="Life expectancy at birth (years)", hue="Country", style="event",
    markers=True, dashes=False
)
plt.show()
```
